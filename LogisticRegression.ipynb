## Feature Normalization 


```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
norm_data = scaler.fit(data.data).transform(data.data)
```

```python
print(norm_data[:10,:])
```

## Train-Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(norm_data, data.target, test_size=0.2, random_state=42)

print('Number of train samples: ', len(X_train))
print('Number of test samples: ', len(X_test))
```

## Logistic Regression Model 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

clf = LogisticRegression().fit(X_train,y_train)
y_test_preds = clf.predict(X_test)

print("Accuracy on test dataset: ", accuracy_score(y_test, y_test_preds))
print("Confusion matrix on test dataset: ")
print(confusion_matrix(y_test, y_test_preds))
```
