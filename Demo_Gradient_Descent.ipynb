```python
# Making the imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0, 9.0)

#create dataset

wo = 0.0
w1 = 1.0

x = np.arange(100)
y = np.random.normal(x*w1 + wo,10)

plt.scatter(x, y)
plt.show()
```

```python
# Building the model
w1_trainable = 0.01
wo_trainable = 0.01
lr = 0.0001  # The learning Rate
epochs = 1000  # The number of iterations to perform gradient descent

n = len(x)# Number of elements in X

# Performing Gradient Descent 
for i in range(epochs): 
    y_pred = w1_trainable*x + wo_trainable  # The current predicted value of Y
    D_w1_trainable = (-2/n) * sum(x * (y - y_pred))  # Derivative wrt m
    D_wo_trainable = (-2/n) * sum(y - y_pred)  # Derivative wrt c
    w1_trainable -= lr * D_w1_trainable  # Update m
    wo_trainable -= lr * D_wo_trainable  # Update c
    
print (w1_trainable, wo_trainable)
```

```python
plt.scatter(x, y)
plt.plot(x,x*w1_trainable+wo_trainable)
plt.show()
```
